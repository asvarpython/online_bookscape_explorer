```python
import pandas as pd
df = pd.read_csv(r'dataset/books_data.csv')
df
```

```python
# Sanity Check of the data
df.shape
```

```python
#info
df.info()
```

```python
#finding missing value
df.isnull().sum()
```

```python
object_columns = ['book_subtitle', 'book_authors', 'book_description', 'industryIdentifiers', 'categories', 'imageLinks', 'currencyCode_listPrice', 'currencyCode_retailPrice', 'buyLink', 'publisher']
df[object_columns] = df[object_columns].fillna('Unknown')
```

```python
float_column = ['year']
df[float_column] = df[float_column].fillna(0)
```

```python
df.isnull().sum()
```

```python
print(df.dtypes)
```

```python
df.to_csv(r'dataset/books_data_cleaned_df.csv', index=False)
```
